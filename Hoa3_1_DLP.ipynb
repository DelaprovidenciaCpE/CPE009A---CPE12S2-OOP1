```python
#Make a program that would calculate the distance between two points given a list of coordinates. Use the distance formula
coordinate_list = [(1,1),(2,3)]
x1 = coordinate_list[0][0]
y1 = coordinate_list[0][1]

x2 = coordinate_list[1][0]
y2 = coordinate_list[1][1]

distance = ((x2-x1)**2 + (y2-y1)**2)**0.5
print("The distance between the two points is:", distance)
```

```python
#For a given string input, replace all the words “stupid” with an asterisk * equal to the length of the string. The newstring value should be displayed with the asterisks.

text = input("Enter a word:")
word = 'stupid'
ast = '*' * len(word)

result = text.replace(word, ast)
result = result.replace(word.capitalize(), ast.capitalize())
result = result.replace(word.upper(), ast.upper())

print(result)
```

```python
#Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a person’s name,conact,and address based from a user input which is the id of the record.
phonebook = [
{"id": 1, "name" : "Dane", "contact" : "0676767676767", "address" : "Qc"},
{"id": 2, "name" : "Angel", "contact" : "0123456789", "address" : "Cavite"},
{"id": 3, "name" : "Mochi", "contact" : "0987654321", "address" : "Manila"},
]

user_id = int(input("Enter the id:"))
found = False

for person in phonebook:
  if person["id"] == user_id:
    print("Name:", person["name"])
    print("Contact:", person["contact"])
    print("Address:", person["address"])
    found = True
    break

if not found:
  print("Person not found")
```
