# For Loop

```python
week = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "sunday"]
for x in week:
  print(x)
```

# **Break Statement**

```python
week = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "sunday"]
for x in week:
  print(x)
  if x == "thursday":
    break


```

# Nested Loops

```python
#Example1
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]
for x in adjective:
  for y in fruits:
    print(x,y)
```

# While Loop

```python
i = 1
while i<6:
  print(i)
  i += 1 #i = i+1 Assignment operator
```

#The break statement while loop

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
   break
  i += 1
```

# The continue statement

```python
i=0
while i<6:
  i+=1
  if i==3:
   print(i)
   continue
```

# Else statement

```python
i = 1
while i<6:
  print(i)
  i+=1
else:
  print("i is no longer less than 6")
```
