Develop a python scripts that:

    Calculate total marks and percentage
    Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

    Determine overall result (Pass/Fail)


Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"


```python

def assign_grade(marks):
    if 90 <= marks <= 100:
        return "A+"
    elif 80 <= marks <= 89:
        return "A"
    elif 70 <= marks <= 79:
        return "B+"
    elif 60 <= marks <= 69:
        return "B"
    elif 50 <= marks <= 59:
        return "C"
    elif 40 <= marks <= 49:
        return "D"
    else:
        return "F"

def performance_remarks(percentage):
    if percentage >= 90:
        return "Outstanding"
    elif percentage >= 80:
        return "Excellent"
    elif percentage >= 70:
        return "Very Good"
    elif percentage >= 60:
        return "Good"
    elif percentage >= 40:
        return "Satisfactory"
    else:
        return "Needs Improvement"

def main():
    subjects = int(input("How many courses:"))
    marks_list = []
    grades = []

    for i in range(subjects):
        marks = int(input(f"Enter grade for subject:{i+1} "))
        marks_list.append(marks)
        grades.append(assign_grade(marks))

    total_marks = sum(marks_list)
    percentage = total_marks / subjects

    overall_result = "Pass" if all(m >= 40 for m in marks_list)  else "Fail"

    print("\nOverall Result")
    for i in range(subjects):
        print(f"Subject {i+1}: Grade = {marks_list[i]}, Remarks = {grades[i]}")

    print(f"Performance Remarks: {performance_remarks(percentage)}")


if __name__ == "__main__":
    main()
```
