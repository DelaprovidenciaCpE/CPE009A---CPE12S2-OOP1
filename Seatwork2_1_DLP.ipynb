```python
# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termCrit(actNumber):
    rawScore = []
    totalScore = []

    for i in range(actNumber):
        while True:
            inputScore = float(input("Score: "))
            inputTotal = float(input("Total: "))

            if inputScore > inputTotal:
                print("\nInvalid score!")
            elif inputScore < 0 or inputTotal <= 0:
                print("\nInvalid input!")
            else:
                rawScore.append(inputScore)
                totalScore.append(inputTotal)
                break

    grade = (sum(rawScore) / sum(totalScore)) * 100
    return grade

```

```python
# Create a function for the class standing per term
def termCS():
    print("HOA:")
    hoa_num = int(input("How many hands-on activities this term? "))
    termHOA = termCrit(hoa_num)

    print("QUIZZES:")
    quiz_num = int(input("How many quizzes this term? "))
    termQuiz = termCrit(quiz_num)

    print("ASSIGNMENTS:")
    ass_num = int(input("How many assignments? "))
    termAss = termCrit(ass_num)

    grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
    return grade
```

```python
# Create a function to compute the term grade
def termGrade():
  print("Computing for the term grade.")
  examRaw = float(input("Exam Score: "))
  examTotal = float(input("Exam Total: "))
  examGrade = (examRaw/examTotal) * 100

  CSgrade = termCS()

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade
```

```python
# Create a function to compute for the grades in the semester
def semesterGrade():
    print("\nPrelim Grade")
    prelimGrade = termGrade()

    print("\nMidterm Grade")
    midtermGrade = termGrade()
    finalMG = (1/3) * prelimGrade + (2/3) * midtermGrade

    print("\nFinal Grade")
    finalGrade = termGrade()
    finalFG = (1/3) * finalMG + (2/3) * finalGrade

    return finalFG
```

```python
def convertGrade(finalGrade):
    if finalGrade >= 97:
        return 1.00
    elif finalGrade >= 94:
        return 1.25
    elif finalGrade >= 91:
        return 1.50
    elif finalGrade >= 88:
        return 1.75
    elif finalGrade >= 85:
        return 2.00
    elif finalGrade >= 82:
        return 2.25
    elif finalGrade >= 79:
        return 2.50
    elif finalGrade >= 76:
        return 2.75
    elif finalGrade >= 75:
        return 3.00
    else:
        return 5.00
        print("Failed")
```

```python
# Driver Code

print("\n#Trial 1 ")
print("Equivalent Grade:", convertGrade(semesterGrade()))

print("\n#Trial 2")
print("Equivalent Grade:", convertGrade(semesterGrade()))

print("\n#Trial 3")
print("Equivalent Grade for a Whole Semester:", convertGrade(semesterGrade()))
```
